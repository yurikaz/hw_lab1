tring_01 = “Mama, just killed a man, put a gun against his head, pulled my trigger, now he's dead. Mama, life had just begun, but now I've gone and thrown it all away. Mama, ooh (any way the wind blows), didn't mean to make you cry. If I'm not back again this time tomorrow carry on, carry on as if nothing really matters.” 

1.Посчитайте, сколько в строке символов (исключая пробелы).
2.Посчитайте, сколько в строке слов.
3.Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки, заканчивающиеся на эту букву (например, “s”).
4.Не обязательно: Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки, начинающиеся на эту букву (наприме m)

```python
string_01 = "Mama, just killed a man, put a gun against his head, pulled my trigger, now he's dead. Mama, life had just begun, but now I've gone and thrown it all away. Mama, ooh (any way the wind blows), didn't mean to make you cry. If I'm not back again this time tomorrow carry on, carry on as if nothing really matters."
```

```python
import re

print('1. В строке {} символов, исключая пробелы'.format(len(re.findall('[^\s]{1}', string_01))))
print('2. В строке {} слов'.format(len(re.findall('[\w]+', string_01))))

def end_letter(findin, letter):
    result = re.findall('[\w]+'+letter, findin)
    return result

def begin_letter(findin, letter):
    result = re.findall(r'\bm[\w]+', findin, re.I)
    return result

print('3. Слова, заканчивающиеся на заданную букву')
print(end_letter(string_01, 's'))

print('4. Слова, начинающиеся на заданную букву')
print(begin_letter(string_01, 'm'))
print('Не повторять одинаковые', list(set(begin_letter(string_01, 'm'))))
```
